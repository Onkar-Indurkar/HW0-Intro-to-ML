```python
df = pd.read_csv('https://raw.githubusercontent.com/Onkar-Indurkar/Intro-to-ML/main/forestfires.csv')
df.head()
```

```python
list1 = ['month']

def binary_map (v1) :
    return v1.map({'jan' : 0, "feb": 1, "mar":2, "apr":3, "may":4, "jun":5, "jul":6, 'aug' : 7, "sep": 8, "oct":9, 'nov' : 10, "dec": 11})

df[list1] = df[list1] . apply(binary_map)
list2 = ['day']

def binary_map (v2) :
    return v2.map({'mon' : 0, "tue": 1, 'wed' : 2, "thu": 3, 'fri' : 4, "sat": 5, 'sun' : 6 })

df[list2] = df[list2] . apply(binary_map)
df.tail()
```

```python
x = df.iloc[:, [4,5,6,7,8,9,10]].values #1st 12 cols
y = df.iloc[:, 12].values   #13th col 


y[y >0 ] = 1
y = y.astype(int)
y = y.reshape(-1,1)
```

```python
from sklearn import preprocessing
from sklearn import utils

lab_enc = preprocessing.LabelEncoder()
encoded = lab_enc.fit_transform(y)
```

```python
#spliting the dataset
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(x, y, train_size=0.8, test_size=0.2, random_state = 0)
```

```python
#scaling/standardizing the dataset
from sklearn.preprocessing import StandardScaler 
sc_X = StandardScaler()

X_train = sc_X.fit_transform(X_train)
X_test = sc_X.transform(X_test)
```

```python
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score

# List of kernel names and C values to try
kernel_list = ['linear', 'poly', 'rbf', 'sigmoid']
C_list = [0.1, 1, 10, 100]

# Iterate over kernel names and C values
for kernel in kernel_list:
    for C in C_list:
        print("Kernel:", kernel)
        print("C:", C)
        
        # Create classifier with current kernel and C value
        classifier = SVC(kernel=kernel, C=C)
        
        # Train classifier and make predictions
        classifier.fit(X_train, Y_train)
        Y_pred = classifier.predict(X_test)
        
        # Evaluate performance
        cnf_matrix = confusion_matrix(Y_test, Y_pred)
        accuracy_rate = accuracy_score(Y_test, Y_pred)
        precision_rate = precision_score(Y_test, Y_pred)
        recall_rate = recall_score(Y_test, Y_pred)
        
        # Print evaluation results
        print("Confusion Matrix:")
        print(cnf_matrix)
        print("Accuracy:", accuracy_rate)
        print("Precision:", precision_rate)
        print("Recall:", recall_rate)
        print("")
```
